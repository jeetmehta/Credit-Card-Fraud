```python
# Import necessary libraries
%matplotlib inline
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
# Read in the data
df = pd.read_csv("creditcard.csv")
```

```python
df.describe() # heavy skew in the data + no missing values
```

```python
# Unbalanced dataset
df.Class.value_counts() 
```

```python
# Visualizing the imbalance
counts = df.Class.value_counts()
counts.plot(kind = "bar")
plt.title("Fraud Cases Histogram")
plt.xlabel("Fraud")
plt.ylabel("Number of Instances")
```

```python
# Undersample dataset to remove imbalance
def undersample(data, ratio = 1):
    minority = data[data.Class == 1]
    majority_undersampled = data[data.Class == 0].sample(n = ratio*len(minority))
    return pd.concat([minority, majority_undersampled]).sample(frac = 1) # shuffle data

# Store original unbalanced data
X = df.drop('Class', axis = 1)
Y = df['Class']

# Store undersampled balanced data
df_undersampled = undersample(df)
X_undersampled = df_undersampled.drop('Class', axis=1)
Y_undersampled = df_undersampled.Class
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import *
from sklearn.linear_model import LogisticRegression

# Logistic Regression - Results aggregated using K-Fold Cross validation (K = 5)
kf = KFold(5)
foldAUPRCScores = []
for train_index, test_index in kf.split(X_undersampled):    
    # Get fold data
    x_train_fold, x_test_fold = X_undersampled.iloc[train_index, :], X_undersampled.iloc[test_index, :]
    y_train_fold, y_test_fold = Y_undersampled.iloc[train_index], Y_undersampled.iloc[test_index]
    
    # Create Logistic Regression Model
    logReg = LogisticRegression(C = 0.01)
    
    # Train model on fold data
    logReg.fit(x_train_fold, y_train_fold)
    
    # Predict values using test set
    y_predicted_fold = logReg.predict(x_test_fold)
    
    # Evaluate performance using the Area Under the Precision-Recall Curve (AUPRC)
    recall = average_precision_score(y_test_fold, y_predicted_fold)
    foldAUPRCScores.append(recall)

averageAUPRC = np.mean(foldAUPRCScores)
print "Average Area Under Precision-Recall Curve (AUPRC) Accuracy: ", averageRecall
```
